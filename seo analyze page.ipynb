```python
def clean_me(html):
    soup = BeautifulSoup(html, "lxml")
    for s in soup(['script', 'style']):
        s.decompose()
    return ' '.join(soup.stripped_strings)

def requestURL(url):
    r = requests.get(url)
    bodytext = clean_me(r.content)
    tree = html.fromstring(r.content)
    soup = BeautifulSoup(r.content, "lxml")
    title = tree.xpath('/html/head/title/text()')
    description = tree.xpath('/html/head/meta[@name="description"]/@content') or ""
    body = soup.find('body')
    h1 = tree.xpath('//h1')[0].text.strip()
    return (title[0], h1, body, bodytext, description[0])

def get_tokens(res):
    text = res
    lowers = text.lower().translate(str.maketrans('','',string.punctuation)).translate(str.maketrans('','','1234567890'))
    tokens = nltk.word_tokenize(lowers)
    return tokens

my_url = "http://www.wikipedia.org"
results = requestURL(my_url)
```
